```python
import pandas as pd

# Melt the DataFrame from wide to long format
test_nominal_gdp = test_nominal_gdp.melt(
    id_vars=['Country Name', 'Indicator Name'],
    value_vars=[str(year) for year in range(1988, 2023)],
    var_name='Year',
    value_name='Nom_GDP_temp'  # Temporary name to avoid conflicts
)

# Convert Year to integer
test_nominal_gdp['Year'] = test_nominal_gdp['Year'].astype(int)

# Rename to final column name if needed
test_nominal_gdp.rename(columns={'Nom_GDP_temp': 'Nom_GDP'}, inplace=True)
test_nominal_gdp = test_nominal_gdp.sort_values(by=["Country Name", "Year"], ascending=[True, True])
```
